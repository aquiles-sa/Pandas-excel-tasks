```python
import pandas as pd

productsDataFrame = pd.read_excel("../data/Products.xlsx")
productsDataFrame
```

```python

# TODO(I) Calculate the total revenue("Quantidade vendida" * "Valor Unitário") for each line
# TODO(II) Find and display the product with the greatest total revenue.

productsDataFrame["Receita Total"] = productsDataFrame["Quantidade Vendida"] * productsDataFrame["Valor Unitário"]

productWithGreatestRevenue = productsDataFrame.loc[productsDataFrame["Receita Total"].idxmax()]
print(productWithGreatestRevenue)
```

```python

# TODO(I) Extract the year and month from "Data da venda" and add these informations as new columns in Dataframe
# TODO(II) Calculate and display the sum of "Quantidade Vendida" for each year and month

productsDataFrame["Data da Venda"] = pd.to_datetime(productsDataFrame["Data da Venda"])

productsDataFrame["Ano"] = productsDataFrame["Data da Venda"].dt.year
productsDataFrame["Mês"] = productsDataFrame["Data da Venda"].dt.month

sumQuantityOfSales = productsDataFrame.groupby(["Ano", "Mês"])["Quantidade Vendida"].sum().reset_index()
sumQuantityOfSales
```
